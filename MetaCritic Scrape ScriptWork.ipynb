```python
containers = page_soup.findAll('div', {'class':'review_section'}) ##we will also need to capture album name
#critic_name_soup = page_soup.findAll('div', {'class':'source'}) #review_critic includes user reviews
#("h2", {"class":"pi-item pi-item-spacing pi-title"}

critic_names= []
critic_scores = []
critic_reviews = []
#("h2", {"class":"pi-item pi-item-spacing pi-title"}
for container in containers: #42 loops

    if re.search('<div class="source">',str(container)) == None : ##only grab critic reviews, not user reviews
        continue
        
    #add artist name
    
    #add album name
    
    critic_name_soup = container.findAll('div', {'class':'source'})
    for name in critic_name_soup:
        #critic_name = re.sub('<[^>]+>', '', str(name))
        critic_name = name.text
        critic_names.append(critic_name)
    
    critic_score_soup = container.findAll('div', {'class':'review_grade'})
    for score in critic_score_soup:
        #critic_score = re.sub('<[^>]+>', '', str(score))
        test_score = re.search('\d+', str(score))
        critic_scores.append(test_score.group(0))
    
    critic_review_soup = container.findAll('div', {'class':'review_body'})
    for review in critic_review_soup:
        #critic_review = re.sub('<[^>]+>', '', str(review))
        #critic_review = re.sub('\n', '', str(critic_review))
        critic_review = review.text.strip()
        critic_reviews.append(critic_review)

print(critic_names)
print(critic_scores)
print(critic_reviews)
```

```python
class AlbumReview:
    def __init__(self, artist, album, reviewer, score, review):
        self.artist = artist
        self.album = album
        self.reviewer = reviewer
        self.score = score
        self.review = review

    def __str__(self):
        return f'Artist: {self.artist}, Album: {self.album}, Reviewer: {self.reviewer}, Score: {self.score}, Review: {self.review}'
    
    def __repr__(self):
        return (f'{self.__class__.__name__}('
           f'{self.artist!r}, {self.album!r}, {self.reviewer!r}, {self.score!r}, {self.review!r} )')
    #def __repr__(self):
     #   return {'artist':self.artist, 'album':self.album, 'reviewer':self.reviewer, 'score':self.score, 'review':self.review}
```

```python
import csv
file_name = 'csv_test'
def save_reviews_to_csv_file(review_list, file_name):
    with open(file_name, 'w', newline='') as csvfile:
        fieldnames = ['artist', 'album','reviewer', 'score', 'review']#list(review_list[0].__dict__.keys())
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for review in review_list:
            writer.writerow({'artist': review.artist, 'album': review.album, 
                             'reviewer':review.reviewer, 'score':review.score, 'review':review.review})
```

```python
file_name =  'review_writing_test.csv'

save_reviews_to_csv_file(review_list,file_name)
```

```python
url = 'https://www.metacritic.com/browse/albums/score/metascore/all/filtered?page=0'
req = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, "html.parser")
#print(page_soup.prettify())
```

```python
for a_tag in soup.findAll("a"):
    href = a_tag.attrs.get("href")
    
    if href == "" or href is None:
    # href empty tag
        continue
        # join the URL if it's relative (not absolute link)
    href = urljoin(url, href)
    parsed_href = urlparse(href)
    # remove URL GET parameters, URL fragments, etc.
    href = parsed_href.scheme + "://" + parsed_href.netloc + parsed_href.path
    #print(href)
    if re.search('critic-reviews',str(href)) != None :
        #print('match')
        print(href)
```
